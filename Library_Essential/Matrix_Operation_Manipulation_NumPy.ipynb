# Task: Use NumPy library to complete the following tasks
1. Create two linear arrays a and b with 20 random elements each ranging from 1 to 100.
2. Now change those linear arrays to two 5x4 matrices.
3. Perform scalar multiplication with matrix a and b.
4. Now try to modify the matrix b and perform matrix multiplication with a and b to store it in
c.
5. Print only the column 2, 3 of matrix c. Print the whole matrix using slicing technique.
6. Find the largest element in the array, and also the index of that element.
7. Convert the matrix c into an one dimensional array keeping the original array
unchanged.

# Solution:

```python
#1
import numpy as np

a = np.random.randint(1, 101, 20)
b = np.random.randint(1, 101, 20)

print(a)
print(b)
```

```python
#2
a = a.reshape(5, 4)
b = b.reshape(5, 4)

print(a)
print(b)
```

```python
#3
scalar_mul_a = a * 2
scalar_mul_b = b * 3

print(scalar_mul_a)
print(scalar_mul_b)
```

```python
#4
b[0, 0] = 78
c = np.dot(a, b.T)
print(c)
```

```python
#5
print(c[:, 1:3])
print(c[:, :])
```

```python
#6
maxm = c.max()
maxm_ind = np.unravel_index(c.argmax(), c.shape)

print(maxm)
print(maxm_ind)
```

```python
#7
c_flat = c.flatten()
print(c_flat)
```

# Task: Use Matplotlib library to complete the following tasks:
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]
1. Generate a scatter plot graph for temperature_dhaka vs humidity_dhaka using (*) as the
marker in color red.
study_hours = [2,3,4,4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]
2. Now do a line plot for study_hours vs marks with figure size (12,8)
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]
3. Now draw a horizontal bar graph showing the scores in each of the subjects.
4. Draw a vertical bar graph where each subject score is visualized in different colors.
5. Show all these graphs in one graph using subplot( ) function.

# Solution:

```python
#1
import matplotlib.pyplot as plt

temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]

plt.figure(figsize=(6, 4))
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.xlabel("temperature_dhaka (°C)")
plt.ylabel("humidity_dhaka (%)")
plt.title("temperature_dhaka vs humidity_dhaka")
plt.grid(True)
plt.show()
```

```python
#2
study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks1 = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

plt.figure(figsize=(12, 8))
plt.plot(study_hours, marks1, marker='o', linestyle='-', color='blue')
plt.xlabel("study_Hours")
plt.ylabel("marks")
plt.title("study_Hours vs marks")
plt.grid(True)
plt.show()
```

```python
#3
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]

plt.figure(figsize=(8, 5))
plt.barh(subjects, marks, color='purple')
plt.xlabel("marks")
plt.ylabel("subjects")
plt.title("Scores in Subjects")
plt.show()
```

```python
#4
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]

colors = ['blue', 'green', 'red', 'orange', 'purple']

plt.figure(figsize=(8, 5))
plt.bar(subjects, marks, color=colors)
plt.xlabel("subjects")
plt.ylabel("marks")
plt.title("Scores in Subjects")
plt.show()
```

```python
#5
fig, axes = plt.subplots(2, 2, figsize=(15, 10))


axes[0, 0].scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
axes[0, 0].set_xlabel("temperature_dhaka (°C)")
axes[0, 0].set_ylabel("humidity_dhaka (%)")
axes[0, 0].set_title("temperature_dhaka vs humidity_dhaka")


axes[0, 1].plot(study_hours, marks1, marker='o', linestyle='-', color='blue')
axes[0, 1].set_xlabel("study_hours")
axes[0, 1].set_ylabel("marks")
axes[0, 1].set_title("study_hours vs marks")


axes[1, 0].barh(subjects, marks, color='purple')
axes[1, 0].set_xlabel("marks")
axes[1, 0].set_ylabel("subjects")
axes[1, 0].set_title("Scores in Subjects")


axes[1, 1].bar(subjects, marks, color=colors)
axes[1, 1].set_xlabel("subjects")
axes[1, 1].set_ylabel("marks")
axes[1, 1].set_title("Scores in Subjects")

plt.tight_layout()
plt.show()
```
